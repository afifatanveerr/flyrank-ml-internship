# 3. Quick look at the data

The dataset contains **30,000 pages** and **44 features** describing historical content performance.

Some observed statistics are:

- Average impressions over the last 90 days: **5200.37**
- Average clicks over the last 90 days: **16.10**
- Average CTR: **0.51**
- **15,581** pages have a CTR greater than 5%.

These measured values show that the dataset contains enough variation to support further analysis of content refresh opportunities.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

url = "https://raw.githubusercontent.com/afifatanveerr/flyrank-ml-internship/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)

print("Dataset Loaded Successfully\n")

print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())

display(df.head())

print("\nAverage Impressions:")
print(df["impressions_90d"].mean())

print("\nAverage Clicks:")
print(df["clicks_90d"].mean())

print("\nAverage CTR:")
print(df["ctr"].mean())

print("\nPages with CTR greater than 5%:")
print((df["ctr"] > 0.05).sum())
```
